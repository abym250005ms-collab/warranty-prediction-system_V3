# Warranty Dataset EDA

This notebook runs the complete EDA pipeline for `warranty_dataset.xlsx` (`Clean_merge_data` sheet) and surfaces core insights for claim prediction and cost forecasting.

```python
from eda.common import ensure_output_dirs, load_dataset, save_json
from eda.eda_data_overview import run_data_overview
from eda.eda_univariate import run_univariate_analysis
from eda.eda_relationships import run_relationship_analysis
from eda.eda_temporal import run_temporal_analysis
from eda.eda_risk_analysis import run_risk_analysis
from eda.eda_visualizations import generate_visualizations
from eda.eda_summary_report import create_summary_report
```

```python
input_file = 'warranty_dataset.xlsx'
sheet_name = 'Clean_merge_data'
output_dir = 'outputs'

paths = ensure_output_dirs(output_dir)
df = load_dataset(input_file, sheet_name)
results = {
    'data_overview': run_data_overview(df, paths['distributions']),
    'univariate': run_univariate_analysis(df, paths['distributions']),
    'relationships': run_relationship_analysis(df, paths['correlations']),
    'temporal': run_temporal_analysis(df, paths['temporal']),
    'risk_analysis': run_risk_analysis(df, paths['risk_analysis']),
    'visualizations': generate_visualizations(df, paths['geographic']),
}
results['summary'] = create_summary_report(results, paths['base'])
save_json(results, 'outputs/eda_summary_statistics.json')
print('EDA complete. See outputs directory for plots and reports.')
```
